Raw counts may not be the most informative statistic. One common improvement is to use *inverse document frequency*, the inverse of the proportion of documents that contain a given term.

````python
# TODO: Compute the number of distinct documents in the collection.
N = 0
doc_df=df["id"].unique().tolist()
N = len(doc_df)
print("Number of distinct documents: ", N)
````

````python
# TODO: Compute the number of distinct documents each term appears in
# and store in a dictionary.
dict_df = dict()
# for loop to iterate all distinct documents
for i in range(N):
  # document id name
  temp_id = df["id"][i]
  # create dataframe with documnet id name
  temp_df = df[df["id"] == temp_id]
  # shape of dataframe to calculate number of term appears
  number_of_docs = temp_df.shape
  # store number in dictionary
  dict_df[temp_id] = number_of_docs[0]
````

````python
# TODO: Print the relative document frequency of 'the',
# i.e., the number of documents that contain 'the' divided by N.

## create dataframe with term "the"
df_the = df[df["term"] == "the"]
# shape of the dataframe (index 0 is number of documents)
shape = df_the.shape
print("Relative document frequency of 'the':", shape[0]/N)
````

Empricially, we usually see better retrieval results if we rescale term frequency (within documents) and inverse document frequency (across documents) with the log function. Let the `tfidf` of term _t_ in document _d_ be:
```
tfidf(t, d) = log(count(t, d) + 1) * log(N / df(t))
```

Later in the course, we will show a probabilistic derivation of this quantity based on smoothing language models.

````python
# TODO: Compute the tf-idf value for each term in each document.
# Take the raw term data and add a tfidf field to each record.
tfidf_terms = None
tf_idf_id_term_dict = dict()
for i in range(N):
  # document id name
  temp_id = df["id"][i]
  # create dataframe with document id name
  temp_df = df[df["id"] == temp_id]
  # word vectors dataframe
  temp_word_vec = temp_df.term.apply(str.split).apply(pd.value_counts).fillna(0)
  # calculate term-frequency
  temp_tf = temp_word_vec.divide(np.sum(temp_word_vec, axis=1), axis=0)
  # inverse document frequency
  temp_idf = np.log10(len(temp_tf) / temp_word_vec[temp_word_vec > 0].count()) 
  # tf-idf
  tfidf_terms = np.multiply(temp_tf, temp_idf.to_frame().T).reset_index(drop=True)
  # terms
  terms = temp_df["term"].unique().tolist()
  for word in terms:
    temp_word_tfidf = tfidf_terms[word].max()
    temp_key = "id: " + temp_id +", term: "+ word
    tf_idf_id_term_dict[temp_key] = temp_word_tfidf
````

````python
tf_idf_id_term_dict.items()
````

````python
ids = []
values = []
for key, value in tf_idf_id_term_dict.items():
  ids.append(key)
  values.append(value)
````

````python
# dictionary of lists 
tfidf_dict_final = {'id:term': ids, 'tf-idf': values}
````

````python
final_tfidf_df = pd.DataFrame(tfidf_dict_final)
final_tfidf_df.head()
````

````python
# TODO: Print the 20 term-document pairs with the highest tf-idf values.
df_20= final_tfidf_df.sort_values(by=['tf-idf'], ascending=False).reset_index(drop=True)
## list out all the terms
term_doc_20 = df_20["id:term"].tolist()

## print top 10 terms
print(term_doc_20[0:20])
````
